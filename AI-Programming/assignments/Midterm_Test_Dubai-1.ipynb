### Question 10:
Imagine you're working with an image represented as a 3D numpy array of shape `(height, width, 3)`, where the last dimension corresponds to the RGB (Red, Green, and Blue) colour channels. You have two 2D arrays:

1. `point_coords`: A 2D array of shape `(n, 2)` containing coordinates of `n` points in the image. Each row is a coordinate `(y, x)`, where `y` is the row index and `x` is the column index in the image array.
2. `colors`: A 2D array of shape `(n, 3)` containing RGB values that you want to assign to each of the `n` points in the image.

Use fancy indexing to set the RGB values of the specified points in the image to the following given RGB values from the `colors` array such that the pixel at (2, 2) becomes red, the pixel at (1, 3) becomes green, and the pixel at (3, 1) becomes blue.

### An example representation of RGB image in computer memory
$
$

image.png

```python
image = np.zeros((3, 5, 5), dtype=np.uint8)  # A 5x5 image with all black pixels
point_coords = np.array([[2, 2],
                        [1, 3],
                        [3, 1]])

colors = np.array([[255, 0, 0],   # Red
                   [0, 255, 0],   # Green
                   [0, 0, 255]])  # Blue

# Your code here
for i in range(len(point_coords)):
    x, y = point_coords[i]
    image[:, x, y] = colors[i]
image
```

### Question 11:
A NumPy array named data_matrix contains information about various products sold by a company over 12 months. Each row in the array represents a different product, and each column corresponds to a month's sales data. The sales data includes the number of units sold (as integers) for each product each month. The company sells 15 different products, and thus the data_matrix is of shape (15, 12). Consider the following data_matrix provided below.

Find the month with the highest sales total. The answer should be represented as integer, where 1 will represent January, 2 for February, and so on.

```python
np.random.seed(0)  # Do not change
data_matrix = np.random.randint(1, 100, (15, 12))

# Your code here
month_with_highest_sales = np.argwhere(data_matrix == data_matrix.max())

month_with_highest_sales
```

```python
# right solution
# Calculate the total sales for each month (column-wise sum)
monthly_totals = np.sum(data_matrix, axis=0)

# Find the month with the highest total sales
highest_month = np.argmax(monthly_totals) + 1  # Adding 1 to represent January as 1

print("Month with the highest sales total:", highest_month)
```
